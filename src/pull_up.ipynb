```python
from Pose import Pose
import cv2
import numpy as np
```

```python
state='neutral'
counter=0
```

```python
p=Pose(static_image_mode=False,model_complexity=1)
```

```python
cap=cv2.VideoCapture('../videos/pull-up.mp4')


state='ok'
counter=0

while cap.isOpened():

    ret,frame=cap.read()
 

    
    
    if not ret:
        break

    
    h,w,_=frame.shape
    left_hand_angle=p.find_angle(frame,[11,13,15])
    right_hand_angle=p.find_angle(frame,[12,14,16])

    angle=(left_hand_angle+right_hand_angle)//2

    

    print(angle)
    # if angle > 100:
    #     state='neutral'
    #     print('ok')
   
   

    if angle  > 130:
        state='DOWN'

        

    if angle <60 and state=='DOWN':
        state='up' 
        counter+=1
        print('up_ok')
        
    cv2.putText(frame, 'STATE:{}'.format(state), (50, 50),
                        cv2.FONT_HERSHEY_PLAIN, 2, (0, 0, 255), 2)
    cv2.putText(frame, 'Counter:{:03d}'.format(counter), (h-15, 50),
                        cv2.FONT_HERSHEY_PLAIN, 2, (0, 0, 255), 2)
    
    cv2.imshow('',frame)
    if cv2.waitKey(1)==27:
        break

    print('{}: {}'.format(state,counter))
cap.release()
cv2.destroyAllWindows()
```
